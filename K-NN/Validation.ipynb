```python
import pandas as pd
import numpy as np
```

```python
attribute = np.array(data)[:, :]
```

```python
#get distance from two coordinate
def getEuclidean(coordinate1,coordinate2):
    distance = 0
    for i in range(len(coordinate2)-1):
        distance += pow((coordinate1[i]-coordinate2[i]),2)
    return pow(distance,0.5)
```

```python
def getNeighbour(dataTest, dataTraining):
    length = len(dataTraining)
    distances = []
    for i in range(length):
        dist = getEuclidean(dataTest, dataTraining[i])
        distances.append((dist, dataTraining[i][-1]))
    
    distances.sort(key = lambda x:x[0])
    return distances
```

```python
def splitDataTest(start, end, data):
    test = []
    
    while(start < end):
        test.append(data[start])
        start+=1
    return test
```

```python
def splitDataTraining(startIndex, endIndex, data):
    dataTraining = []
    
    if(startIndex == 0):
        temp = data[endIndex:]
        for i in range(len(temp)):
            dataTraining.append(temp[i])
            
    elif(endIndex == len(data)):
        temp = data[:startIndex]
        for i in range(len(temp)):
            dataTraining.append(temp[i])
            
    else:
        temp = data[:startIndex]
        for i in range(len(temp)):
            dataTraining.append(temp[i])
        
        temp = data[endIndex:]
        for i in range(len(temp)):
            dataTraining.append(temp[i])
        
    return dataTraining
```

```python
k = 10
lengthEksperimen = 100 // k
lengthData = len(attribute) * k // 100 
startIndex = 0
endIndex = lengthData
dataTest = []
dataTrain = []

for i in range(lengthEksperimen):
    dataTest.append(splitDataTest(startIndex, endIndex, attribute))
    dataTrain.append(splitDataTraining(startIndex, endIndex, attribute))
    
    startIndex+=lengthData
    endIndex+=lengthData
```

```python
len(dataTrain[0])
```

```python
neighbour=[]
for i in range(len(dataTest)):
    for j in range(len(dataTest[i])):
        temp = getNeighbour(dataTest[i][j], dataTrain[i])
        neighbour.append(temp)
```

```python
neighbour[0]
```

```python
def separateByClass(attribute):
    separated = {}
    for i in range(len(attribute)):
        vector = attribute[i]
        if (vector[-1] not in separated):
            separated[vector[-1]] = []
        separated[vector[-1]].append(vector)
    
    copy = list(separated)
    return copy
```

```python
getClass = separateByClass(attribute)
```

```python
getClass
```

```python
total = 0
total2 = 0
total3 = 0
hasil = []

for i in range(len(neighbour)):
    for j in range(k):
        if(neighbour[i][j][-1] == getClass[0]):
            total += 1            
        elif(neighbour[i][j][-1] == getClass[1]):
            total2 += 1
        elif(neighbour[i][j][-1] == getClass[2]):
            total3 += 1
    
    if(total >= total2 and total >= total3):
        hasil.append((getClass[0], total))
    elif(total2 >= total and total2 >= total3):
        hasil.append((getClass[1], total2))
    elif(total3 >= total and total3 >= total2):
        hasil.append((getClass[2], total3))
    
    total = 0
    total2 = 0
    total3 = 0
```

```python
hasil[0]
```

```python
startIndex = 0
endIndex = lengthData
splitResult = []

for i in range(lengthEksperimen):
    splitResult.append(splitDataTest(startIndex, endIndex, hasil))
    startIndex+=lengthData
    endIndex+=lengthData
```

```python
error = 0
hasilError = []

for i in range(lengthEksperimen):
    for j in range(lengthData):
        if(dataTest[i][j][-1] != splitResult[i][j][0]):
            error +=1
    message = "Hasil Error :"
    hasilError.append(error/len(attribute)*100)
    error = 0
    
```
