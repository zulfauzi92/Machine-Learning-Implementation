```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Iris.csv')
data
```

```python
atributte = np.array(data)[:, :]
```

```python
atributte
```

```python
#get distance from two coordinate
def getEuclidean(coordinate1,coordinate2):
    distance = pow((coordinate1[0]-coordinate2[0]),2)+ pow((coordinate1[0]-coordinate2[0]),2)
    return pow(distance,0.5)
```

```python
def getNeighbour(dataTest, dataTraining):
    length = len(dataTraining)-1
    distances = []
    for i in range(length):
        dist = getEuclidean(dataTest, dataTraining[i])
        distances.append((dist,atributte[i]))
    
    distances.sort(key = lambda x:x[0])
    return distances
```

```python
def splitDataSet(dataTest, k):
    dataTraining = []
    for i in range(k):
        if((dataTest == atributte[i]).all()):
            continue
        else:
            dataTraining.append(atributte[i][:-1])
    return dataTraining
```

```python
k = 140
result = []
test = []

for i in range(k):
    dataTest = atributte[i]
    dataTraining = splitDataSet(dataTest, k)
    b = getNeighbour(dataTest, dataTraining)
    result.append(b)
    test.append(dataTest)
```
